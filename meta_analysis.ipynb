```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o dataset
#file_path = "seu_arquivo.csv"  # Substitua pelo caminho do seu arquivo
#data = pd.read_csv(file_path)
data = pd.read_csv('/content/drive/MyDrive/machine learning/Gráficos/meta-analysis_SMOTE.csv', sep=',', encoding='iso-8859-1')

# Inspecionar os dados
print(data.head())

# Diferença relativa (AFTER - BEFORE)
data['DELTA'] = data['AFTER'] - data['BEFORE'] #((data['AFTER'] / data['BEFORE']) - 1)


# Análise estatística: diferença média por ML e métrica
delta_stats = data.groupby(['ML', 'CODE_SMELL', 'METRIC'])['DELTA'].mean().reset_index()
print(delta_stats)

# Visualização das diferenças (DELTA)
plt.figure(figsize=(12, 6))
sns.barplot(data=delta_stats, x="METRIC", y="DELTA", hue="CODE_SMELL", ci=None)
#plt.title("Mean Difference (AFTER - BEFORE) by Code Smell and Metric - SMOTE")
plt.xlabel("Metric")
plt.ylabel("Mean Difference (AFTER - BEFORE)")
plt.legend(title="Code Smell")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Análise estatística: diferença média por ML e métrica
delta_stats = data.groupby(['ML', 'CODE_SMELL', 'METRIC'])['DELTA'].mean().reset_index()
print(delta_stats)

# Visualização das diferenças (DELTA)
plt.figure(figsize=(12, 8))
sns.barplot(data=delta_stats, x="METRIC", y="DELTA", hue="ML", ci=None)
#plt.title("Mean Difference (AFTER - BEFORE) by ML and Metric - SMOTE")
plt.xlabel("Metric", fontsize=20)
plt.ylabel("Mean Difference (AFTER - BEFORE)", fontsize=20)
plt.legend(fontsize=20)
plt.xticks(rotation=45, fontsize=20)
plt.yticks(fontsize=20)
plt.tight_layout()
plt.show()

```

# **6.3. Analysis of the average impact between different preprocessing techniques (ANOVA)**

```python
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Carregar o CSV em um DataFrame
data = pd.read_csv('/content/drive/MyDrive/machine learning/Gráficos/meta-analysis.csv', sep=',', encoding='iso-8859-1')

print(data['TECHNIQUE'].nunique())
print(data['ML'].nunique())
print(data['CODE_SMELL'].nunique())
print(data['METRIC'].nunique())
print(len(data))

pd.crosstab([data['TECHNIQUE'], data['ML']], [data['CODE_SMELL'], data['METRIC']])
```

```python
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Carregar o CSV em um DataFrame
data = pd.read_csv('/content/drive/MyDrive/machine learning/Gráficos/meta-analysis.csv', sep=',', encoding='iso-8859-1')

# ANOVA para BEFORE
model_before = ols('BEFORE ~ C(TECHNIQUE) + C(ML):C(TECHNIQUE) + C(TECHNIQUE):C(CODE_SMELL) + C(TECHNIQUE):C(METRIC) + C(ML):C(TECHNIQUE):C(CODE_SMELL) + C(ML):C(TECHNIQUE):C(METRIC) + C(TECHNIQUE):C(CODE_SMELL):C(METRIC)' , data=data).fit()
anova_results_before = anova_lm(model_before, typ=2)

# ANOVA para AFTER
model_after = ols('AFTER ~ C(TECHNIQUE) + C(ML):C(TECHNIQUE) + C(TECHNIQUE):C(CODE_SMELL) + C(TECHNIQUE):C(METRIC) + C(ML):C(TECHNIQUE):C(CODE_SMELL) + C(ML):C(TECHNIQUE):C(METRIC) + C(TECHNIQUE):C(CODE_SMELL):C(METRIC) ', data=data).fit()
anova_results_after = anova_lm(model_after, typ=2)

# Mostrar os resultados em formato de tabela
print("ANOVA Resultados para BEFORE:")
print(anova_results_before)
print("\nANOVA Resultados para AFTER:")
print(anova_results_after)
```
